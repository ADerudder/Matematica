```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=y=np.linspace(-10,10,200)
xv,yv=np.meshgrid(x,y)
```

```python
a=2
b=3
x=y=np.linspace(-10,10,200)
xv,yv=np.meshgrid(x,y)
f=a*xv**2 + b*yv**2
plt.figure(figsize=(9,6))
plt.contour(xv,yv,f,cmap='inferno')
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$p_x$',fontsize=12)
plt.title("Oscilador armonico")
plt.colorbar()
```

```python
a=3
b=3
x=y=np.linspace(0,10,200)
xv,yv=np.meshgrid(x,y)
f=a/xv + (yv**2)*b
plt.figure(figsize=(9,6))
plt.contour(xv,yv,f,cmap='inferno',levels=10)
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$r$',fontsize=12)
plt.ylabel('$P_r$',fontsize=12)
plt.title("Potencial Kepler")
plt.colorbar()
```

```python
x=y=np.linspace(-8,8,200)
xv,yv=np.meshgrid(x,y)
a=2
b=3
f=a*np.cos(xv) + yv**2 * 1/b
plt.figure(figsize=(9,6))
plt.contour(xv,yv,f,cmap='inferno',levels=30)
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$  $',fontsize=12)
plt.ylabel('$P_$',fontsize=12)
plt.title("Pendulo simple")
plt.colorbar()
```

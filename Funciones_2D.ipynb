```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=y=np.linspace(-10,10,200)
xv,yv=np.meshgrid(x,y)
```

```python
f=xv**2 + yv**2
```

```python
plt.pcolormesh(xv,yv,f,cmap='inferno')
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$y$',fontsize=12)
plt.colorbar()
```

```python
g=np.sin(xv)*yv

plt.pcolormesh(xv,yv,g,cmap='Greens')
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$y$',fontsize=12)
plt.colorbar()
```

Le ponemos una condicion:

$$ x +y^2 < 20 $$ para $(x,y)=$ [-10,10] x [-10,10] 

```python
h=(xv + yv**2 <20).astype(float) # h es un vector que contiene  donde x+y**2 <20 y x+y**2 donde x+y**2>20
plt.pcolormesh(xv,yv,h,cmap='Spectral')
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$y$',fontsize=12)
plt.colorbar()
```

```python
a=2
b=3
f=a*np.cos(xv) + yv**2 * 1/b
plt.contour(xv,yv,f,cmap='inferno')
plt.grid('on', linestyle = 'dashed', alpha = 0.5)
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$y$',fontsize=12)
plt.colorbar()
```
